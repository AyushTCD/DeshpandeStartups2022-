# Quarter wise Location



```python
q1df = smalldf[smalldf['Month']<4]
q2df = smalldf[(smalldf['Month']>3) & (smalldf['Month']<7)]
q3df = smalldf[(smalldf['Month']>6) & (smalldf['Month']<10)]
q4df = smalldf[smalldf['Month']>9]
```

```python
q1LC = q1df['City  Location'].value_counts().head(10)
q2LC = q2df['City  Location'].value_counts().head(10)
q3LC = q3df['City  Location'].value_counts().head(10)
q4LC = q4df['City  Location'].value_counts().head(10)
```

```python
fig = plt.figure(figsize=(10,10))

plt.subplot(2, 2, 1)
q1LC.plot(kind='bar')
plt.title("Q1")
plt.xlabel("Location")
plt.ylabel("Frequencies")


plt.subplot(2, 2, 2)
q2LC.plot(kind='bar')
plt.title("Q2")
plt.xlabel("Location")
plt.ylabel("Frequencies")

plt.subplot(2, 2, 3)
q3LC.plot(kind='bar')
plt.title("Q3")
plt.xlabel("Location")
plt.ylabel("Frequencies")


plt.subplot(2, 2, 4)
q4LC.plot(kind='bar')
plt.title("Q4")
plt.xlabel("Location")
plt.ylabel("Frequencies")


plt.tight_layout()
plt.show()
```

```python
q1IV = q1df['Industry Vertical'].value_counts().head(10)
q2IV = q2df['Industry Vertical'].value_counts().head(10)
q3IV = q3df['Industry Vertical'].value_counts().head(10)
q4IV = q4df['Industry Vertical'].value_counts().head(10)


fig = plt.figure(figsize=(10,10))

plt.subplot(2, 2, 1)
q1IV.plot(kind='bar')
plt.title("Q1")
plt.xlabel("Industries")
plt.ylabel("Frequencies")


plt.subplot(2, 2, 2)
q2IV.plot(kind='bar')
plt.title("Q2")
plt.xlabel("Industries")
plt.ylabel("Frequencies")

plt.subplot(2, 2, 3)
q3IV.plot(kind='bar')
plt.title("Q3")
plt.xlabel("Industries")
plt.ylabel("Frequencies")


plt.subplot(2, 2, 4)
q4IV.plot(kind='bar')
plt.title("Q4")
plt.xlabel("Industries")
plt.ylabel("Frequencies")


plt.tight_layout()
plt.show()
```
